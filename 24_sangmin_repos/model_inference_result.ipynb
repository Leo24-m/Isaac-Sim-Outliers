### Visuallize isaac sim IMU sensor data

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 경로
file_path = 'C:/Users/RECS419/ssource/repos/24_sangmin_repos/isaac_imu_4.csv'

# CSV 파일 읽기
df_isaac = pd.read_csv(file_path, header=0)

plt.figure(figsize=(20, 5))

# 각 열에 대한 그래프 생성
plt.plot(df_isaac.index, df_isaac.iloc[:, 1], label='lin_acc_x')  # 첫 번째 열
plt.plot(df_isaac.index, df_isaac.iloc[:, 2], label='lin_acc_y')  # 두 번째 열
plt.plot(df_isaac.index, df_isaac.iloc[:, 3], label='lin_acc_z')  # 세 번째 열
plt.plot(df_isaac.index, df_isaac.iloc[:, 4], label='ang_vel_x')  # 첫 번째 열
plt.plot(df_isaac.index, df_isaac.iloc[:, 5], label='ang_vel_y')  # 두 번째 열
plt.plot(df_isaac.index, df_isaac.iloc[:, 6], label='ang_vel_z')  # 세 번째 열

# 그래프 제목과 축 레이블 설정
plt.title('Isaac Sim Accel XYZ value Results', fontsize=20)
plt.xlabel('Time', fontsize=16)
plt.ylabel('Value', fontsize=16)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# 범례 표시
plt.legend(fontsize=14)
# 그래프 출력
plt.show()
```
