### Visuallize isaac sim IMU sensor data

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV  
file_path = 'C:/Users/RECS419/ssource/repos/24_sangmin_repos/isaac_imu_4.csv'

# CSV  
df_isaac = pd.read_csv(file_path, header=0)

plt.figure(figsize=(20, 5))

#     
plt.plot(df_isaac.index, df_isaac.iloc[:, 1], label='lin_acc_x')  #   
plt.plot(df_isaac.index, df_isaac.iloc[:, 2], label='lin_acc_y')  #   
plt.plot(df_isaac.index, df_isaac.iloc[:, 3], label='lin_acc_z')  #   
plt.plot(df_isaac.index, df_isaac.iloc[:, 4], label='ang_vel_x')  #   
plt.plot(df_isaac.index, df_isaac.iloc[:, 5], label='ang_vel_y')  #   
plt.plot(df_isaac.index, df_isaac.iloc[:, 6], label='ang_vel_z')  #   

#     
plt.title('Isaac Sim Accel XYZ value Results', fontsize=20)
plt.xlabel('Time', fontsize=16)
plt.ylabel('Value', fontsize=16)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

#  
plt.legend(fontsize=14)
#  
plt.show()
```
